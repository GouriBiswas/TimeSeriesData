```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

url = "https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv"
df = pd.read_csv(url, parse_dates=['Date'], index_col='Date')

df_global = df.groupby('Date').sum()

print(df_global.head())

plt.figure(figsize=(12, 6))
plt.plot(df_global['Confirmed'], marker='o', linestyle='-', color='b', label='Total Confirmed Cases')
plt.title('Time Series Data of COVID-19 Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.legend()
plt.grid()
plt.show()

df_global.fillna(method='ffill', inplace=True)

decomposition = seasonal_decompose(df_global['Confirmed'], model='multiplicative', period=30)
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()

rolling_mean = df_global['Confirmed'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(df_global['Confirmed'], label='Original Data')
plt.plot(rolling_mean, color='red', label='30-Day Moving Average')
plt.title('Moving Average on COVID-19 Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.legend()
plt.show()

```
